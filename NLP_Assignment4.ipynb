Build a Named Entity Recognition (NER) system for extracting
entities from real-world text such as news articles or social media
data. And measure its accuracy, precision, recall, and F1 score.

Dataset

```python
texts = [
"Narendra Modi visited New York to attend the UN General Assembly.",
"Apple launched the iPhone 15 in California on Tuesday.",
"Virat Kohli scored a century against Australia in Mumbai."
]
```

```python
true_entities = [
[("Narendra Modi", "PERSON"), ("New York", "GPE"), ("UnGeneral Assembly", "ORG")],
[("Apple", "ORG"), ("iPhone 15", "PRODUCT"), ("California", "GPE"), ("Tuesday", "DATE")],
[("Virat Kohli", "PERSON"), ("Australia", "GPE"), ("Mumbai", "GPE")]
]
```

Install & Load Model

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

Extract Predicted Entities

```python
predicted_entities = []

for text in texts:
  doc = nlp(text)
  predicted_entities.append([(ent.text, ent.label_) for ent in doc.ents])

print(predicted_entities)
```

Convert Entities to Token Labels

```python
def get_token_labels(text, entities):
  labels = ["O"] * len(text.split())
  for ent_text, ent_label in entities:
    ent_tokens = ent_text.split()
    text_tokens = text.split()

    for i in range(len(text_tokens)):
      if text_tokens[i:i+len(ent_tokens)] == ent_tokens:
        for j in range(len(ent_tokens)):
          labels[i+j] = ent_label

  return labels
```

Generate True & Predicted Labels

```python
y_true = []
y_pred = []
```

```python
for i, text in enumerate(texts):
  y_true.extend(get_token_labels(text, true_entities[i]))
  y_pred.extend(get_token_labels(text, predicted_entities[i]))
```

Compute Metrics

```python
from sklearn.metrics import accuracy_score, classification_report

print("Accuracy:", accuracy_score(y_true, y_pred))
print("\nClassification Report:\n")
print(classification_report(y_true, y_pred))
```
