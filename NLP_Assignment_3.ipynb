Import Libraries

```python
import pandas as pd
import re
import nltk

from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.preprocessing import LabelEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

Raw Text

```python
data = {
    'text': [
        "I love Natural Language Processing!",
        "NLP is amazing and powerful.",
        "I hate spam messages."
    ],
    'label': ['positive', 'positive', 'negative']
}

df = pd.DataFrame(data)
```

Text Cleaning Function

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-z\s]', '', text)
    return text
```

Lemmatization and Stop Word Removal

```python
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

def preprocess(text):
    words = text.split()
    words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    return " ".join(words)

df['cleaned_text'] = df['text'].apply(clean_text)
df['final_text'] = df['cleaned_text'].apply(preprocess)
```

Label Encoding

```python
encoder = LabelEncoder()
df['encoded_label'] = encoder.fit_transform(df['label'])
```

TF-IDF Vectorization

```python
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(df['final_text'])

tfidf_df = pd.DataFrame(
    tfidf_matrix.toarray(),
    columns=vectorizer.get_feature_names_out()
)
```

Save Outputs

```python
df.to_csv("processed_text.csv", index=False)
tfidf_df.to_csv("tfidf_features.csv", index=False)
```
